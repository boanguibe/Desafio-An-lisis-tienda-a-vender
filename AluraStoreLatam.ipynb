### Importaci√≥n de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. An√°lisis de facturaci√≥n



Inspeccionar los datos (opcional pero recomendado)
Esto nos permite ver qu√© columnas contiene el DataFrame y c√≥mo est√°n estructuradas.

```python
print(tienda.columns)
print(tienda.head())
```

‚úÖ Calcular el ingreso total por tienda

```python
# Ingreso total por tienda = suma de la columna "Precio"
ingreso_tienda1 = tienda["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()

# Mostrar los resultados
print("Ingreso total Tienda 1:", ingreso_tienda1)
print("Ingreso total Tienda 2:", ingreso_tienda2)
print("Ingreso total Tienda 3:", ingreso_tienda3)
print("Ingreso total Tienda 4:", ingreso_tienda4)

```

üìò Explicaci√≥n:
tienda["Precio"] accede a la columna de precios del DataFrame.

.sum() suma todos los valores de esa columna, es decir, el total facturado.

# 2. Ventas por categor√≠a

‚úÖ ¬øQu√© haremos?
Usar groupby sobre la columna "Categor√≠a del Producto".

Aplicar .size() o .count() para contar la cantidad de productos vendidos por categor√≠a.

Ordenar los resultados para ver las categor√≠as m√°s populares.

```python
# Ventas por categor√≠a para Tienda 1
ventas_cat_tienda1 = tienda.groupby("Categor√≠a del Producto").size().sort_values(ascending=False)

# Tienda 2
ventas_cat_tienda2 = tienda2.groupby("Categor√≠a del Producto").size().sort_values(ascending=False)

# Tienda 3
ventas_cat_tienda3 = tienda3.groupby("Categor√≠a del Producto").size().sort_values(ascending=False)

# Tienda 4
ventas_cat_tienda4 = tienda4.groupby("Categor√≠a del Producto").size().sort_values(ascending=False)

# Mostrar los resultados
print("Tienda 1 - Ventas por categor√≠a:\n", ventas_cat_tienda1, "\n")
print("Tienda 2 - Ventas por categor√≠a:\n", ventas_cat_tienda2, "\n")
print("Tienda 3 - Ventas por categor√≠a:\n", ventas_cat_tienda3, "\n")
print("Tienda 4 - Ventas por categor√≠a:\n", ventas_cat_tienda4)
```

üìò Explicaci√≥n:
groupby("Categor√≠a del Producto") agrupa los datos por categor√≠a.

.size() cuenta cu√°ntas veces aparece cada categor√≠a.

.sort_values(ascending=False) ordena de mayor a menor para ver las m√°s vendidas primero.

üß© Extra: Unir todo en un DataFrame para comparar entre tiendas

```python
ventas_categorias = pd.DataFrame({
    "Tienda 1": ventas_cat_tienda1,
    "Tienda 2": ventas_cat_tienda2,
    "Tienda 3": ventas_cat_tienda3,
    "Tienda 4": ventas_cat_tienda4
}).fillna(0).astype(int)

print(ventas_categorias)
```

# 3. Calificaci√≥n promedio de la tienda


üéØ Objetivo
Calcular el promedio de la columna Calificaci√≥n para cada tienda. Esta columna refleja la puntuaci√≥n que el cliente le da a su experiencia o al producto (de 1 a 5, seg√∫n los datos).

```python
# Calcular la valoraci√≥n media por tienda
valoracion_tienda1 = tienda["Calificaci√≥n"].mean()
valoracion_tienda2 = tienda2["Calificaci√≥n"].mean()
valoracion_tienda3 = tienda3["Calificaci√≥n"].mean()
valoracion_tienda4 = tienda4["Calificaci√≥n"].mean()

# Mostrar los resultados
print("Valoraci√≥n media Tienda 1:", round(valoracion_tienda1, 2))
print("Valoraci√≥n media Tienda 2:", round(valoracion_tienda2, 2))
print("Valoraci√≥n media Tienda 3:", round(valoracion_tienda3, 2))
print("Valoraci√≥n media Tienda 4:", round(valoracion_tienda4, 2))
```

üìò Explicaci√≥n:
.mean() calcula el promedio de las calificaciones.

round(valor, 2) redondea a 2 decimales para una presentaci√≥n m√°s clara.

üß© **Extra: Resumen en DataFrame**

```python
valoraciones = pd.DataFrame({
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Valoraci√≥n Media": [valoracion_tienda1, valoracion_tienda2, valoracion_tienda3, valoracion_tienda4]
})

print(valoraciones.round(2))
```

# 4. Productos m√°s y menos vendidos

üìä Productos m√°s vendidos y menos vendidos en cada tienda
üéØ Objetivo
Identificar qu√© productos se vendieron m√°s veces y cu√°les se vendieron menos veces en cada tienda. Como cada fila representa una venta, basta con contar cu√°ntas veces aparece cada producto en la columna "Producto".

```python
# Tienda 1
ventas_productos_t1 = tienda["Producto"].value_counts()
mas_vendido_t1 = ventas_productos_t1.head(1)
menos_vendido_t1 = ventas_productos_t1.tail(1)

# Tienda 2
ventas_productos_t2 = tienda2["Producto"].value_counts()
mas_vendido_t2 = ventas_productos_t2.head(1)
menos_vendido_t2 = ventas_productos_t2.tail(1)

# Tienda 3
ventas_productos_t3 = tienda3["Producto"].value_counts()
mas_vendido_t3 = ventas_productos_t3.head(1)
menos_vendido_t3 = ventas_productos_t3.tail(1)

# Tienda 4
ventas_productos_t4 = tienda4["Producto"].value_counts()
mas_vendido_t4 = ventas_productos_t4.head(1)
menos_vendido_t4 = ventas_productos_t4.tail(1)

# Mostrar resultados
print("üõçÔ∏è Tienda 1\nM√°s vendido:", mas_vendido_t1, "\nMenos vendido:", menos_vendido_t1, "\n")
print("üõçÔ∏è Tienda 2\nM√°s vendido:", mas_vendido_t2, "\nMenos vendido:", menos_vendido_t2, "\n")
print("üõçÔ∏è Tienda 3\nM√°s vendido:", mas_vendido_t3, "\nMenos vendido:", menos_vendido_t3, "\n")
print("üõçÔ∏è Tienda 4\nM√°s vendido:", mas_vendido_t4, "\nMenos vendido:", menos_vendido_t4, "\n")
```

üìò Explicaci√≥n:
value_counts() cuenta cu√°ntas veces se vendi√≥ cada producto.

.head(1) devuelve el producto m√°s vendido.

.tail(1) devuelve el menos vendido.

üìä (Opcional) Visualizaci√≥n con gr√°fico de barras de los 5 m√°s vendidos por tienda

---


Si quieres ver los 5 productos m√°s vendidos por tienda de forma visual:

```python
import matplotlib.pyplot as plt

# Ejemplo para Tienda 1
ventas_productos_t1.head(5).plot(kind='barh', color='skyblue')
plt.title("Top 5 productos m√°s vendidos - Tienda 1")
plt.xlabel("Cantidad de Ventas")
plt.ylabel("Producto")
plt.gca().invert_yaxis()
plt.show()
```

# 5. Env√≠o promedio por tienda

üöö Valor del env√≠o promedio por tienda
üéØ Objetivo
Calcular el promedio de la columna Costo de env√≠o para cada tienda, lo que nos permitir√° entender cu√°nto cuesta en promedio el env√≠o desde cada tienda.

```python
# Calcular el promedio del costo de env√≠o para cada tienda
envio_prom_t1 = tienda["Costo de env√≠o"].mean()
envio_prom_t2 = tienda2["Costo de env√≠o"].mean()
envio_prom_t3 = tienda3["Costo de env√≠o"].mean()
envio_prom_t4 = tienda4["Costo de env√≠o"].mean()

# Mostrar los resultados redondeados
print("Costo de env√≠o promedio - Tienda 1:", round(envio_prom_t1, 2))
print("Costo de env√≠o promedio - Tienda 2:", round(envio_prom_t2, 2))
print("Costo de env√≠o promedio - Tienda 3:", round(envio_prom_t3, 2))
print("Costo de env√≠o promedio - Tienda 4:", round(envio_prom_t4, 2))
```

üìò Explicaci√≥n:
.mean() calcula el promedio de una columna num√©rica.

round(..., 2) redondea el resultado a 2 decimales para mejor presentaci√≥n.

üìä (Opcional) Tabla resumen con pandas:

```python
envio_promedios = pd.DataFrame({
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Costo de Env√≠o Promedio": [
        envio_prom_t1, envio_prom_t2, envio_prom_t3, envio_prom_t4
    ]
})

print(envio_promedios.round(2))
```

üìä 1. Gr√°fico de Barras: Ingreso total por tienda
Mostraremos cu√°nto ingreso gener√≥ cada tienda.

```python
import matplotlib.pyplot as plt

ingresos = [tienda["Precio"].sum(), tienda2["Precio"].sum(), tienda3["Precio"].sum(), tienda4["Precio"].sum()]
nombres_tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

plt.figure(figsize=(8, 5))
plt.bar(nombres_tiendas, ingresos, color='skyblue')
plt.title("Ingreso Total por Tienda")
plt.ylabel("Ingresos ($)")
plt.xlabel("Tiendas")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

üìà 2. Gr√°fico de L√≠neas: Valor promedio del env√≠o por tienda
Representa el costo promedio de env√≠o por tienda.

```python
envios_prom = [
    tienda["Costo de env√≠o"].mean(),
    tienda2["Costo de env√≠o"].mean(),
    tienda3["Costo de env√≠o"].mean(),
    tienda4["Costo de env√≠o"].mean()
]

plt.figure(figsize=(8, 5))
plt.plot(nombres_tiendas, envios_prom, marker='o', linestyle='-', color='green')
plt.title("Costo Promedio de Env√≠o por Tienda")
plt.ylabel("Costo de Env√≠o ($)")
plt.xlabel("Tiendas")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
