### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



Inspeccionar los datos (opcional pero recomendado)
Esto nos permite ver qué columnas contiene el DataFrame y cómo están estructuradas.

```python
print(tienda.columns)
print(tienda.head())
```

✅ Calcular el ingreso total por tienda

```python
# Ingreso total por tienda = suma de la columna "Precio"
ingreso_tienda1 = tienda["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()

# Mostrar los resultados
print("Ingreso total Tienda 1:", ingreso_tienda1)
print("Ingreso total Tienda 2:", ingreso_tienda2)
print("Ingreso total Tienda 3:", ingreso_tienda3)
print("Ingreso total Tienda 4:", ingreso_tienda4)

```

📘 Explicación:
tienda["Precio"] accede a la columna de precios del DataFrame.

.sum() suma todos los valores de esa columna, es decir, el total facturado.

# 2. Ventas por categoría

✅ ¿Qué haremos?
Usar groupby sobre la columna "Categoría del Producto".

Aplicar .size() o .count() para contar la cantidad de productos vendidos por categoría.

Ordenar los resultados para ver las categorías más populares.

```python
# Ventas por categoría para Tienda 1
ventas_cat_tienda1 = tienda.groupby("Categoría del Producto").size().sort_values(ascending=False)

# Tienda 2
ventas_cat_tienda2 = tienda2.groupby("Categoría del Producto").size().sort_values(ascending=False)

# Tienda 3
ventas_cat_tienda3 = tienda3.groupby("Categoría del Producto").size().sort_values(ascending=False)

# Tienda 4
ventas_cat_tienda4 = tienda4.groupby("Categoría del Producto").size().sort_values(ascending=False)

# Mostrar los resultados
print("Tienda 1 - Ventas por categoría:\n", ventas_cat_tienda1, "\n")
print("Tienda 2 - Ventas por categoría:\n", ventas_cat_tienda2, "\n")
print("Tienda 3 - Ventas por categoría:\n", ventas_cat_tienda3, "\n")
print("Tienda 4 - Ventas por categoría:\n", ventas_cat_tienda4)
```

📘 Explicación:
groupby("Categoría del Producto") agrupa los datos por categoría.

.size() cuenta cuántas veces aparece cada categoría.

.sort_values(ascending=False) ordena de mayor a menor para ver las más vendidas primero.

🧩 Extra: Unir todo en un DataFrame para comparar entre tiendas

```python
ventas_categorias = pd.DataFrame({
    "Tienda 1": ventas_cat_tienda1,
    "Tienda 2": ventas_cat_tienda2,
    "Tienda 3": ventas_cat_tienda3,
    "Tienda 4": ventas_cat_tienda4
}).fillna(0).astype(int)

print(ventas_categorias)
```

# 3. Calificación promedio de la tienda


🎯 Objetivo
Calcular el promedio de la columna Calificación para cada tienda. Esta columna refleja la puntuación que el cliente le da a su experiencia o al producto (de 1 a 5, según los datos).

```python
# Calcular la valoración media por tienda
valoracion_tienda1 = tienda["Calificación"].mean()
valoracion_tienda2 = tienda2["Calificación"].mean()
valoracion_tienda3 = tienda3["Calificación"].mean()
valoracion_tienda4 = tienda4["Calificación"].mean()

# Mostrar los resultados
print("Valoración media Tienda 1:", round(valoracion_tienda1, 2))
print("Valoración media Tienda 2:", round(valoracion_tienda2, 2))
print("Valoración media Tienda 3:", round(valoracion_tienda3, 2))
print("Valoración media Tienda 4:", round(valoracion_tienda4, 2))
```

📘 Explicación:
.mean() calcula el promedio de las calificaciones.

round(valor, 2) redondea a 2 decimales para una presentación más clara.

🧩 **Extra: Resumen en DataFrame**

```python
valoraciones = pd.DataFrame({
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Valoración Media": [valoracion_tienda1, valoracion_tienda2, valoracion_tienda3, valoracion_tienda4]
})

print(valoraciones.round(2))
```

# 4. Productos más y menos vendidos

📊 Productos más vendidos y menos vendidos en cada tienda
🎯 Objetivo
Identificar qué productos se vendieron más veces y cuáles se vendieron menos veces en cada tienda. Como cada fila representa una venta, basta con contar cuántas veces aparece cada producto en la columna "Producto".

```python
# Tienda 1
ventas_productos_t1 = tienda["Producto"].value_counts()
mas_vendido_t1 = ventas_productos_t1.head(1)
menos_vendido_t1 = ventas_productos_t1.tail(1)

# Tienda 2
ventas_productos_t2 = tienda2["Producto"].value_counts()
mas_vendido_t2 = ventas_productos_t2.head(1)
menos_vendido_t2 = ventas_productos_t2.tail(1)

# Tienda 3
ventas_productos_t3 = tienda3["Producto"].value_counts()
mas_vendido_t3 = ventas_productos_t3.head(1)
menos_vendido_t3 = ventas_productos_t3.tail(1)

# Tienda 4
ventas_productos_t4 = tienda4["Producto"].value_counts()
mas_vendido_t4 = ventas_productos_t4.head(1)
menos_vendido_t4 = ventas_productos_t4.tail(1)

# Mostrar resultados
print("🛍️ Tienda 1\nMás vendido:", mas_vendido_t1, "\nMenos vendido:", menos_vendido_t1, "\n")
print("🛍️ Tienda 2\nMás vendido:", mas_vendido_t2, "\nMenos vendido:", menos_vendido_t2, "\n")
print("🛍️ Tienda 3\nMás vendido:", mas_vendido_t3, "\nMenos vendido:", menos_vendido_t3, "\n")
print("🛍️ Tienda 4\nMás vendido:", mas_vendido_t4, "\nMenos vendido:", menos_vendido_t4, "\n")
```

📘 Explicación:
value_counts() cuenta cuántas veces se vendió cada producto.

.head(1) devuelve el producto más vendido.

.tail(1) devuelve el menos vendido.

📊 (Opcional) Visualización con gráfico de barras de los 5 más vendidos por tienda

---


Si quieres ver los 5 productos más vendidos por tienda de forma visual:

```python
import matplotlib.pyplot as plt

# Ejemplo para Tienda 1
ventas_productos_t1.head(5).plot(kind='barh', color='skyblue')
plt.title("Top 5 productos más vendidos - Tienda 1")
plt.xlabel("Cantidad de Ventas")
plt.ylabel("Producto")
plt.gca().invert_yaxis()
plt.show()
```

# 5. Envío promedio por tienda

🚚 Valor del envío promedio por tienda
🎯 Objetivo
Calcular el promedio de la columna Costo de envío para cada tienda, lo que nos permitirá entender cuánto cuesta en promedio el envío desde cada tienda.

```python
# Calcular el promedio del costo de envío para cada tienda
envio_prom_t1 = tienda["Costo de envío"].mean()
envio_prom_t2 = tienda2["Costo de envío"].mean()
envio_prom_t3 = tienda3["Costo de envío"].mean()
envio_prom_t4 = tienda4["Costo de envío"].mean()

# Mostrar los resultados redondeados
print("Costo de envío promedio - Tienda 1:", round(envio_prom_t1, 2))
print("Costo de envío promedio - Tienda 2:", round(envio_prom_t2, 2))
print("Costo de envío promedio - Tienda 3:", round(envio_prom_t3, 2))
print("Costo de envío promedio - Tienda 4:", round(envio_prom_t4, 2))
```

📘 Explicación:
.mean() calcula el promedio de una columna numérica.

round(..., 2) redondea el resultado a 2 decimales para mejor presentación.

📊 (Opcional) Tabla resumen con pandas:

```python
envio_promedios = pd.DataFrame({
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Costo de Envío Promedio": [
        envio_prom_t1, envio_prom_t2, envio_prom_t3, envio_prom_t4
    ]
})

print(envio_promedios.round(2))
```

📊 1. Gráfico de Barras: Ingreso total por tienda
Mostraremos cuánto ingreso generó cada tienda.

```python
import matplotlib.pyplot as plt

ingresos = [tienda["Precio"].sum(), tienda2["Precio"].sum(), tienda3["Precio"].sum(), tienda4["Precio"].sum()]
nombres_tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

plt.figure(figsize=(8, 5))
plt.bar(nombres_tiendas, ingresos, color='skyblue')
plt.title("Ingreso Total por Tienda")
plt.ylabel("Ingresos ($)")
plt.xlabel("Tiendas")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

📈 2. Gráfico de Líneas: Valor promedio del envío por tienda
Representa el costo promedio de envío por tienda.

```python
envios_prom = [
    tienda["Costo de envío"].mean(),
    tienda2["Costo de envío"].mean(),
    tienda3["Costo de envío"].mean(),
    tienda4["Costo de envío"].mean()
]

plt.figure(figsize=(8, 5))
plt.plot(nombres_tiendas, envios_prom, marker='o', linestyle='-', color='green')
plt.title("Costo Promedio de Envío por Tienda")
plt.ylabel("Costo de Envío ($)")
plt.xlabel("Tiendas")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

🥧 3. Gráfico de Torta: Distribución de productos por categoría (Tienda 1 como ejemplo)
Mostraremos la proporción de ventas por categoría en la tienda 1.

```python
categoria_t1 = tienda["Categoría del Producto"].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(categoria_t1, labels=categoria_t1.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title("Distribución de Categorías (Tienda 1)")
plt.axis('equal')  # Hace el gráfico circular
plt.tight_layout()
plt.show()
```
